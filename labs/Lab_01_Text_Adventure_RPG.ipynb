## Part 2: Building the Character System 

Let's start with the foundation: characters! Every RPG needs a way to represent the player and enemies.

### 2.1 The Character Base Class

Create a `Character` class with:
- **Attributes**: `name`, `health`, `max_health`, `strength`, `defense`
- **Methods**:
  - `__init__()`: Initialize the character
  - `is_alive()`: Returns True if health > 0
  - `take_damage(amount)`: Reduce health (but don't go below 0)
  - `attack(target)`: Roll dice, calculate damage, apply to target
  - `__str__()`: Return a nice string representation

### The D20 Combat System

We'll use a simple dice-based combat system:

1. **Attack Roll**: Roll a d20 (1-20). Add attacker's strength.
2. **Defense Check**: If attack roll > target's defense, hit! Otherwise, miss.
3. **Damage**: On a hit, deal `strength` damage (or roll dice for variable damage)

```python
import random

def roll_d20():
    return random.randint(1, 20)

def roll_dice(num_dice, sides):
    """Roll multiple dice and return the total. E.g., roll_dice(2, 6) for 2d6."""
    return sum(random.randint(1, sides) for _ in range(num_dice))
```

````python
# YOUR CODE: Build your Character base class here
import random

def roll_d20():
    """Roll a 20-sided die."""
    return random.randint(1, 20)

def roll_dice(num_dice, sides):
    """Roll multiple dice and return the total. E.g., roll_dice(2, 6) for 2d6."""
    return sum(random.randint(1, sides) for _ in range(num_dice))


class Character:
    """Base class for all characters in the game."""
    
    def __init__(self, name, health, strength, defense):
        self.name = name
        self.max_health = health
        self.health = health
        self.strength = strength
        self.defense = defense
        pass
    
    def is_alive(self):
        # TODO: Return True if health > 0
        return self.health > 0
    
    def take_damage(self, amount):
        # TODO: Reduce health, but don't go below 0
        # Print a message about the damage taken
        self.health = max(0, self.health - amount)
        print(f"{self.name} takes {amount} damage!")
    
    def attack(self, target):
        # TODO: Implement d20 combat
        # 1. Roll d20, add strength
        # 2. Compare to target's defense
        # 3. If hit, deal damage to target
        # 4. Print combat messages!
        roll = roll_d20()
        total = roll + self.strength
        print(f"{self.name} rolls {roll} + {self.strength} = {total}")
        if total > target.defense:
            damage = max(1, roll_dice(1, 6) + self.strength - target.defense)
            print(f"{self.name} hits for {damage} damage!")
            target.take_damage(damage)
        else:
            print(f"{self.name} misses!")
    
    def __str__(self):
        return f"{self.name} (HP: {self.health}/{self.max_health})"
````

## Part 3: Building the World

Now let's create locations for your adventure!

### 3.1 The Location Class

Each location needs:
- **Attributes**: `name`, `description`, `connections` (dict mapping directions to other locations), `enemies` (list), `items` (list)
- **Methods**:
  - `describe()`: Print the location description and what's here
  - `get_exits()`: Return available directions
  - `add_connection(direction, location)`: Connect to another location
  
**Pro tip**: Store connections as a dictionary: `{"north": cave, "south": village}`

````python
# YOUR CODE: Build the Location class

class Location:
    """A location in the game world."""
    
    def __init__(self, name, description,connections=None, enemies=None, items=None):
        self.name = name
        self.description = description
        self.connections = connections if connections is not None else {}  # {"north": Location, "south": Location, etc.}
        self.enemies = enemies if enemies is not None else []      # List of enemies in this location
        self.items = items if items is not None else []        # List of items in this location
    
    def describe(self):
        """Print a full description of the location."""
        print(f"\n{'='*50}")
        print(f"📍 {self.name}")
        print(f"{'='*50}")
        print(self.description)
        
        # TODO: Show enemies if present
        # TODO: Show items if present
        if self.items:
            print("\nItems here:")
            for item in self.items:
                print(f"- {item}")
        
        # TODO: Show available exits
        exits = self.get_exits()
        if exits:
            print(f"\nAvailable exits: {', '.join(exits)}")
        
        # Show enemies if present
        if self.enemies:
            print("\nEnemies here:")
            for enemy in self.enemies:
                print(f"- {enemy.name} (HP: {enemy.health})")
        
        if self.items:
            print("\nItems here:")
            for item in self.items:
                print(f"- {item}")

    
    def get_exits(self):
        """Return a list of available directions."""
        return list(self.connections.keys())
    
    def add_connection(self, direction, location):
        """Connect this location to another."""
        self.connections[direction] = location
````

## Part 4: The Combat System

Time to make battles exciting! We'll implement a turn-based combat system using a simple state machine.

### 4.1 Combat States

Combat flows through states. For example, here is one simple way to do it:
1. **PLAYER_TURN**: Player chooses an action
2. **ENEMY_TURN**: Enemy attacks
3. **COMBAT_END**: Someone won (or ran away)

```
┌─────────────────────────────────────────────────────────┐
│                                                         │
│  [PLAYER_TURN] ─── attack ──→ [ENEMY_TURN]              │
│       │                            │                    │
│       │                            │                    │
│       │←─────── (enemy alive) ─────┘                    │
│       │                            │                    │
│       │                            ▼                    │
│       │                      (enemy dead?)              │
│       │                            │                    │
│       └── run away ─────→ [COMBAT_END] ←────────────────┘
│                                                         │
└─────────────────────────────────────────────────────────┘
```

````python
# YOUR CODE: Implement the combat system

class Combat:
    """Manages turn-based combat between player and enemy."""
    
    # Combat states
    PLAYER_TURN = "player_turn"
    ENEMY_TURN = "enemy_turn"
    COMBAT_END = "combat_end"
    
    def __init__(self, player, enemy):
        self.player = player
        self.enemy = enemy
        self.state = Combat.PLAYER_TURN
        self.combat_log = []
    
    def start(self):
        """Begin combat and run until someone wins/loses/flees."""
        print(f"\n⚔️ COMBAT BEGINS! ⚔️")
        print(f"{self.player.name} vs {self.enemy.name}!")
        
        while self.state != Combat.COMBAT_END:
            if self.state == Combat.PLAYER_TURN:
                self.player_turn()
            elif self.state == Combat.ENEMY_TURN:
                self.enemy_turn()
        
        return self.get_result()
    
    def player_turn(self):
        """Handle player's turn in combat."""
        print(f"\n{self.player} | {self.enemy}")
        print("What do you do? (attack / run)")
        
        action = input("> ").lower().strip()
        
        if action == "attack":
            self.player.attack(self.enemy)
            if not self.enemy.is_alive():
                print(f"\n🎉 {self.enemy.name} has been defeated!")
                self.state = Combat.COMBAT_END
            else:
                self.state = Combat.ENEMY_TURN
        
        elif action == "run":
            # 50% chance to escape
            if random.random() < 0.5:
                print("You successfully fled!")
                self.state = Combat.COMBAT_END
            else:
                print("Couldn't escape!")
                self.state = Combat.ENEMY_TURN
        
        else:
            print("Invalid action. Try 'attack' or 'run'.")
    
    def enemy_turn(self):
        """Handle enemy's turn in combat."""
        print(f"\n{self.enemy.name}'s turn...")
        self.enemy.attack(self.player)
        
        if not self.player.is_alive():
            print(f"\n💀 {self.player.name} has fallen!")
            self.state = Combat.COMBAT_END
        else:
            self.state = Combat.PLAYER_TURN
    
    def get_result(self):
        """Return the combat result: 'victory', 'defeat', or 'fled'."""
        if not self.enemy.is_alive():
            return "victory"
        elif not self.player.is_alive():
            return "defeat"
        else:
            return "fled"
````

## Part 5: The Main Game Loop

Now let's tie it all together! The `Game` class manages the overall game state and the main loop.

### 5.1 Game States

Your main game should have these states:
- **EXPLORING**: Player is moving around, examining locations
- **IN_COMBAT**: Player is fighting an enemy
- **GAME_OVER**: Player died
- **VICTORY**: Player won!

### 5.2 Available Actions

When exploring, the player might:
- Move: `go north`, `go south`, `go east`, `go west`
- Look around: `look`
- Fight: `fight` or `attack`
- Inventory: `inventory` or `i`
- Get help: `help`
- Quit: `quit`

Handle invalid input gracefully!

````python
# YOUR CODE: Build the main Game class

class Game:
    """Main game controller."""
    
    # Game states
    EXPLORING = "exploring"
    IN_COMBAT = "in_combat"
    GAME_OVER = "game_over"
    VICTORY = "victory"
    
    def __init__(self):
        self.player = None
        self.current_location = None
        self.state = Game.EXPLORING
        self.game_running = True
    
    def start(self):
        """Initialize and start the game."""
        self.show_intro()
        self.current_location = create_world()  # Your function from earlier
        self.current_location.describe()
        self.create_player(self.current_location)
        
        # Main game loop
        while self.game_running:
            if self.state == Game.EXPLORING:
                self.exploration_loop()
            elif self.state == Game.GAME_OVER:
                self.show_game_over()
                break
            elif self.state == Game.VICTORY:
                self.show_victory()
                break
    
    def show_intro(self):
        """Display the game introduction."""
        print("\n" + "="*60)
        print("         YOUR GAME TITLE HERE")
        print("="*60)
        print("\nYour epic intro text goes here...")
        print("Set the scene! What's happening? Why is the player here?")
        print("\n" + "="*60)
    
    def create_player(self, location=None):
        """Create the player character."""
        print("\nWhat is your name, adventurer?")
        name = input("> ")
        self.player = Player(name)
        self.player.location = location if location else self.current_location
        print(f"\nWelcome, {name}! Your adventure begins...")
    
    def exploration_loop(self):
        """Handle player input during exploration."""
        print("\nWhat do you do? (type 'help' for commands)")
        command = input("> ").lower().strip()
        
        # Parse the command
        parts = command.split()
        if not parts:
            return
        
        action = parts[0]
        
        if action == "help":
            self.show_help()
        
        elif action == "look":
            self.current_location.describe()
        
        elif action == "go" and len(parts) > 1:
            direction = parts[1]
            self.move(direction)
        
        elif action in ["north", "south", "east", "west", "up", "down"]:
            self.move(action)
        
        elif action in ["fight", "attack"]:
            self.initiate_combat()
        
        elif action in ["inventory", "i"]:
            self.player.show_inventory()
        
        elif action == "quit":
            print("Thanks for playing!")
            self.game_running = False
            
        elif action == "pickup" and len(parts) > 1:
            item_name = " ".join(parts[1:])
            if item_name.lower() in [item.lower() for item in self.current_location.items]:
                self.player.pick_up(item_name)
                self.current_location.items.remove(item_name)
                print(f"You picked up: {item_name}")
            else:
                print(f"There is no {item_name} here.")
        
        else:
            print("I don't understand that command. Type 'help' for options.")
    
    def move(self, direction):
        """Move the player in the specified direction."""
        if direction in self.current_location.connections:
            self.current_location = self.current_location.connections[direction]
            self.current_location.describe()
            # TODO: Check for automatic combat triggers?
        else:
            print(f"You can't go {direction} from here.")
    
    def initiate_combat(self):
        """Start combat with an enemy in the current location."""
        if not self.current_location.enemies:
            print("There's nothing to fight here.")
            return
        
        enemy = self.current_location.enemies[0]  # Fight first enemy
        battle = Combat(self.player, enemy)
        result = battle.start()
        
        if result == "victory":
            self.current_location.enemies.remove(enemy)
            # TODO: Check for victory condition (e.g., boss defeated)
        elif result == "defeat":
            self.state = Game.GAME_OVER
    
    def show_help(self):
        """Display available commands."""
        print("\n📜 AVAILABLE COMMANDS:")
        print("  go [direction] - Move in a direction (north, south, east, west)")
        print("  look          - Examine your surroundings")
        print("  fight         - Attack an enemy in this location")
        print("  inventory     - Check your inventory")
        print("  help          - Show this help message")
        print("  quit          - Exit the game")
        print("  pickup [item] - Pick up an item in the location")
    
    def show_game_over(self):
        """Display game over message."""
        print("\n" + "="*60)
        print("                    GAME OVER")
        print("="*60)
        print("\nYou have fallen. The adventure ends here...")
        print("\n(But you can always try again!)")
    
    def show_victory(self):
        """Display victory message."""
        print("\n" + "="*60)
        print("                    🎉 VICTORY! 🎉")
        print("="*60)
        print("\nCongratulations! You have completed your quest!")
        print("\nYour name will go down in legend as a hero of the realm!")
        print("\nThank you for playing!")
        # TODO: Add your custom victory text
````

## Final Tips for Success 🎯

1. **Build iteratively**: Get the simplest version working first, then add features.

2. **Test constantly**: After each change, run your code to catch bugs early.

3. **Use print statements**: When debugging, add `print()` statements to see what's happening.

4. **Keep methods small**: Each method should do ONE thing. If it's getting long, break it up.

5. **Have fun**: This is YOUR game. Make it something you enjoy building!
